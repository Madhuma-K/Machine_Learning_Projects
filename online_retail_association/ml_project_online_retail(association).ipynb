```python
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from apyori import apriori
df=pd.read_csv(r"C:\Users\Madhuma k\Downloads\archive (25)\online_retail.csv")
df.head()
```

```python
df=df.dropna(subset=['Description'])
```

```python
df['CustomerID'].fillna('Unknown',inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
le=LabelEncoder()
column_to_convert=['Description','Country']
for col in column_to_convert:
    df[col]=le.fit_transform(df[col])
df.head(10)
```

```python
#records=[]
#for i in range(0,541909):
    #records.append([str(df.values[i,j])for j in range (0,8)])

records=df.iloc[:,0:8].astype(str).values.tolist()
association_rules=apriori(records,min_support=0.0045,min_confidence=0.2,min_lift=3,min_length=2)
association_rules=list(association_rules)
print(association_rules[0])
association_result=list(association_rules)
```
